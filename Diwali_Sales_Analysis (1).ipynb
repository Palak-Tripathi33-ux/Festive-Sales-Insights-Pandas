# 🪔 Diwali Sales Analysis

An exploratory data analysis (EDA) project on Diwali sales data to understand customer buying behavior — who buys, what they buy, and where they're from.

**Goal:** Identify the key customer segments and product categories driving the most sales, to help target marketing & inventory decisions for future Diwali sales.

---


## 1. Import Libraries

```python
# import python libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  # visualizing data
import seaborn as sns

%matplotlib inline
sns.set_style("darkgrid")
```

## 2. Load the Dataset

```python
# import csv file
df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
df.shape
```

```python
df.head()
```

```python
df.info()
```

## 3. Data Cleaning

```python
# drop unrelated/blank columns (Status & unnamed1 have 0 non-null values)
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
# check for null values
pd.isnull(df).sum()
```

```python
# drop rows with null values (only 12 rows affected, out of 11251)
df.dropna(inplace=True)
df.shape
```

```python
# change Amount to integer type for cleaner analysis
df['Amount'] = df['Amount'].astype('int')
df['Amount'].dtypes
```

```python
df.columns
```

```python
# rename column for clarity
df.rename(columns={'Marital_Status': 'Marriage_Status'}, inplace=True)
df.head()
```

## 4. Descriptive Statistics

```python
# describe() method returns description of the data in the DataFrame (count, mean, std, etc.)
df.describe()
```

```python
# describe() for specific columns of interest
df[['Age', 'Orders', 'Amount']].describe()
```

---
## 5. Exploratory Data Analysis

### 5.1 Gender


```python
# plotting a bar chart for Gender and its count
ax = sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Number of Buyers by Gender')
plt.show()
```

```python
# plotting a bar chart for gender vs total amount
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender', y='Amount', data=sales_gen)
plt.title('Total Sales Amount by Gender')
plt.show()
```

> **Insight:** Most buyers are female, and the purchasing power (total amount spent) of women is also significantly higher than men.

### 5.2 Age Group

```python
ax = sns.countplot(data=df, x='Age Group', hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Buyer Count by Age Group and Gender')
plt.show()
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data=sales_age)
plt.title('Total Sales Amount by Age Group')
plt.show()
```

> **Insight:** The majority of buyers fall in the 26-35 age group, with females again leading in both count and total spend.

### 5.3 State

```python
# total number of orders from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

plt.figure(figsize=(15, 5))
sns.barplot(data=sales_state, x='State', y='Orders')
plt.title('Top 10 States by Number of Orders')
plt.xticks(rotation=20)
plt.show()
```

```python
# total amount/sales from top 10 states
sales_state_amt = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

plt.figure(figsize=(15, 5))
sns.barplot(data=sales_state_amt, x='State', y='Amount')
plt.title('Top 10 States by Total Sales Amount')
plt.xticks(rotation=20)
plt.show()
```

> **Insight:** Uttar Pradesh, Maharashtra, and Karnataka lead both in order volume and total sales amount.

### 5.4 Marital Status

```python
plt.figure(figsize=(7, 5))
ax = sns.countplot(data=df, x='Marriage_Status')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Buyer Count by Marital Status (0 = Unmarried, 1 = Married)')
plt.show()
```

```python
sales_marriage = df.groupby(['Marriage_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

plt.figure(figsize=(6, 5))
sns.barplot(data=sales_marriage, x='Marriage_Status', y='Amount', hue='Gender')
plt.title('Total Sales Amount by Marital Status & Gender')
plt.show()
```

> **Insight:** Married women have the highest purchasing power among all customer segments.

### 5.5 Occupation

```python
plt.figure(figsize=(20, 5))
ax = sns.countplot(data=df, x='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Buyer Count by Occupation')
plt.xticks(rotation=30)
plt.show()
```

```python
sales_occ = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

plt.figure(figsize=(20, 5))
sns.barplot(data=sales_occ, x='Occupation', y='Amount')
plt.title('Total Sales Amount by Occupation')
plt.xticks(rotation=30)
plt.show()
```

> **Insight:** Buyers working in the IT, Healthcare, and Aviation sectors contribute the most to total sales.

### 5.6 Product Category

```python
plt.figure(figsize=(20, 5))
ax = sns.countplot(data=df, x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Number of Orders by Product Category')
plt.xticks(rotation=30)
plt.show()
```

```python
sales_cat = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

plt.figure(figsize=(20, 5))
sns.barplot(data=sales_cat, x='Product_Category', y='Amount')
plt.title('Top 10 Product Categories by Total Sales Amount')
plt.xticks(rotation=30)
plt.show()
```

> **Insight:** Food, Clothing & Apparel, and Electronics & Gadgets are the top-selling product categories.

### 5.7 Top Products

```python
# top 10 most ordered products
top_products = df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False)

plt.figure(figsize=(12, 7))
top_products.plot(kind='bar', color=sns.color_palette('viridis', 10))
plt.title('Top 10 Best-Selling Products (by Orders)')
plt.xlabel('Product ID')
plt.ylabel('Total Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

---
## 6. Conclusion

Based on the analysis, the ideal target customer profile for Diwali sales campaigns is:

> **Married women, aged 26-35, from Uttar Pradesh, Maharashtra, and Karnataka, working in IT, Healthcare, or Aviation sectors — most likely to purchase from the Food, Clothing & Apparel, and Electronics & Gadgets categories.**

### Key Takeaways
- 🧑‍🤝‍🧑 **Gender:** Women are both the majority of buyers and have higher purchasing power.
- 🎂 **Age:** The 26-35 age group drives the most sales.
- 📍 **Geography:** UP, Maharashtra, and Karnataka are the top revenue-generating states.
- 💍 **Marital Status:** Married customers (especially women) spend more.
- 💼 **Occupation:** IT, Healthcare, and Aviation professionals are the biggest spenders.
- 🛍️ **Products:** Food, Clothing, and Electronics are the best-selling categories.

These insights can help businesses optimize marketing campaigns, inventory planning, and regional promotions ahead of the next Diwali sales season.
